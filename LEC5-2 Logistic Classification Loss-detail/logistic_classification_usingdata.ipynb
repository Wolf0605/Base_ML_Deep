```python
import tensorflow as tf
import numpy as np
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/content/diabetes.csv')
data.head(2)
```

```python
data = np.asarray(data, dtype='float64')
```

```python
X_data = data[:, :-1]
y_data = data[:, -1]
print(X_data.shape, y_data.shape)
```

```python
model = keras.Sequential(keras.layers.Dense(units=1, input_dim = X_data.shape[1], activation='sigmoid'))
model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=0.0001), metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(X_data, y_data, epochs=500)
```

```python
print("Accuracy: {0}".format(hist.history['accuracy'][-1]))
```

```python
y_predict = model.predict([[0.176471, 0.155779, 0, 0, 0, 0.052161, -0.952178, -0.733333]])
print("Predict: {0}".format(y_predict))
```

```python
evaluate = model.evaluate(X_data, y_data)
print("loss: {0}, accuracy: {1}".format(evaluate[0], evaluate[1]))
```
