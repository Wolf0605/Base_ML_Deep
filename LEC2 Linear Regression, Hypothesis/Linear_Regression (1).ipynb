```python
X_trian = [1,2,3,4,5,6,7,8]
y_train = [10, 20, 30, 40,50,60,70,80]



model = Sequential([keras.layers.Dense(units=1, activation='linear', input_shape=(1,))])

# What's optimizer ??? 
# What's SGD??

# What's compile ?? 

model.compile(loss='mse', optimizer='sgd') 

# prints summary of model
model.summary()
```

```python
model.fit(X_trian, y_train, epochs=100) # What means model.fit? - > It means train data using model

y_predict = model.predict([9,10]) # [5,4] ?? --> if x_test = [5,4] 라는 뜻, , , What's predict - > Litterarly predict [5,4] using model that we already put
# 만약 predict(9) 라고 하면, list index out of range 에러 뜸
```

```python
print(y_predict)

y_predict = model.predict(np.array([9,10]))
print(y_predict)

y_predict = model.predict(np.array([11,13]))
print(y_predict)
```
